```python

import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset("tips")
df
```

```python
sns.lineplot(x="total_bill",y="tip",data=df,hue="sex",linestyle='solid',legend='auto')
```

```python
avg_total_bill = df.groupby('day')['total_bill'].mean()
avg_tip = df.groupby('day')['tip'].mean()
```

```python
plt.figure(figsize=(8,6))
p1 = plt.bar(avg_total_bill.index, avg_total_bill, label='Total Bill')
p2 = plt.bar(avg_tip.index,avg_tip,bottom=avg_total_bill,label='Tip')
plt.xlabel('Day of the week')
plt.ylabel("amount")
plt.title("avgrage total bill and tip by day")
plt.legend()
```

```python
avg_total_bill = df.groupby('time')['total_bill'].mean()
avg_tip = df.groupby('time')['tip'].mean()
```

```python
p1 = plt.bar(avg_total_bill.index, avg_total_bill, label='Total Bill',width=0.4)
p2 = plt.bar(avg_tip.index,avg_tip,bottom=avg_total_bill,label='Tip',width=0.4)
plt.xlabel('Time of the Day')
plt.ylabel("amount")
plt.title("avgrage total bill and tip by time of day")
plt.legend()
```

```python
sns.barplot(x='day',y='total_bill',hue='sex',data=df,palette='Set3')
plt.xlabel('day of week')
plt.ylabel("total bill")
plt.title("Total bill by day and gender")
```

```python
sns.scatterplot(x='total_bill',y='tip',hue='sex',data=df)
plt.xlabel('Total Bill')
plt.ylabel('Tip Amount')
plt.title('Scatter Plot of Total Bill vs. Tip Amt')
```

```python
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
np.random.seed(1)
num_var = np.random.randn(1000)
num_var = pd.Series(num_var,name = "Numerical Variable")
num_var
```

```python
sns.histplot(data = num_var, kde=True)
```

```python
np.random.seed(0)
marks = np.random.normal(loc=70,scale=10,size=100)
marks
```

```python
sns.histplot(data=marks, bins=10,kde=True,stat='count',cumulative=False,multiple='stack',element='bars',palette='Set1',shrink=0.7)
plt.xlabel('Marks')
plt.ylabel('Density')
plt.title('Histogram of Students Marks')
```

```python
sns.boxplot(x=df['day'],y=df['total_bill'],hue=df['sex'])
```

```python
sns.boxplot(x="day", y="total_bill", hue="smoker", data=df, linewidth=2, width=0.6,boxprops={"facecolor": "yellow", "edgecolor": "orange"}, whiskerprops={"color": "black", "linestyle": "--", "linewidth": 1.5}, capprops={"color": "black", "linestyle": "--", "linewidth": 1.5})
```

```python
sns.violinplot(x="day", y="total_bill", hue="smoker", data = df, linewidth=2, width=0.6, palette="Set3", inner="quartile")

#Add labels and title

plt.xlabel("Day of the Week")
plt.ylabel("Total Bill")
plt.title("Violin Plot of Total Bill by Day and Smoker Status")
```
